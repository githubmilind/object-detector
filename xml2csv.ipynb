```python
import os
import xml.etree.ElementTree as Xet
import pandas as pd
import shutil
```

```python
folder_path = 'C:/FAU/bottle_detector/images/OceanPlasticBottleXML/' 

img_source = 'C:/FAU/bottle_detector/images/OceanPlasticBottles/' 
img_dest = 'C:/FAU/bottle_detector/images/bottle-images/' 

out_file = 'bottle_bdnbox.csv'

cols = ["image_name", "x_min", "y_min", "x_max", "y_max", "class_name"]
rows = []

files = os.listdir(folder_path)
i=0
for file in files:
    if i > 100:
        break
    
    if os.path.isfile(os.path.join(folder_path, file)):
        xmlparse = Xet.parse(os.path.join(folder_path, file))
        root = xmlparse.getroot()
        for objects in root.findall('object'):
            xmin = objects.find('bndbox').find('xmin').text
            ymin = objects.find('bndbox').find('ymin').text
            xmax = objects.find('bndbox').find('xmax').text
            ymax = objects.find('bndbox').find('ymax').text
            
            rows.append({"image_name": '/content/drive/MyDrive/object-search-project/bottles/bottle-images/' \
                         + 'img' + str(i) + '.jpg',
                 "x_min": xmin,
                 "y_min": ymin,
                 "x_max": xmax,
                 "y_max": ymax,
                "class_name": "bottle"})
            
            shutil.copyfile(os.path.join(img_source, os.path.splitext(file)[0] + '.jpg'), os.path.join(img_dest, 'img' + str(i) + '.jpg'))
            
            print('img' + str(i) + '.jpg', xmin, ymin, xmax, ymax)
    i = i+1
            
df = pd.DataFrame(rows, columns=cols)

df.to_csv(os.path.join(folder_path, out_file))

```
